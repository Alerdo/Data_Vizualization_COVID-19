### Step 1: Understanding COVID-19's Initial Impact in 2020

To better understand the nature of COVID-19, I began with some dat analysis about the first year of the pandemic.

**Spread Rate**:  
I created a lineplot to illustrate the monthly percentage of the population that tested positive throughout 2020, providing insights into the virus's spread rate.

**Age Groups at Risk**:  
I then developed a bar plot that shows the percentage of COVID-19 deaths by age group with the overall death percentage, illustrating the pandemic's severity but also showing the age groups higher at risk.

**Regions at Risk**:  
A horizontal bar plot was used to display regions with the highest positivity rates(relative to their population sizes), pinpointing areas with significant case numbers.


```python
# Load the data

df_1a = pd.read_excel('DataSet/2020_rates.xlsx', sheet_name='1a')

print(df_1a['Estimated average % of the population that had COVID-19'].dtypes)

# Here I am converting the records that got the % symbol into floats
df_1a['Estimated average % of the population that had COVID-19'] = df_1a['Estimated average % of the population that had COVID-19'] * 100
df_1a['95% Confidence/ Credible Interva Lower'] = df_1a['95% Confidence/ Credible Interva Lower']* 100
df_1a['95% Confidence/ Credible Interval Upper'] = df_1a['95% Confidence/ Credible Interval Upper']* 100



df_1a['Date Time Period'] = pd.to_datetime(df_1a['Time Period'].str.split(' to ').str[1] + ' 2020', format='%d %B %Y')
df_1a['Month Year'] = df_1a['Date Time Period'].dt.strftime('%B')


df_1a.head(40)

```

```python


plt.figure(figsize=(8, 5))
ax = sns.lineplot(data=df_1a, x='Date Time Period', y='Estimated average % of the population that had COVID-19', label="Estimated Average")

# Set the date format on the x-axis
date_format = mdates.DateFormatter('%B')  # Editing the date, choosing just the month 
ax.xaxis.set_major_formatter(date_format)

# Add a band for the confidence interval
plt.fill_between(
    x=df_1a['Date Time Period'],
    y1=df_1a['95% Confidence/ Credible Interva Lower'],
    y2=df_1a['95% Confidence/ Credible Interval Upper'],
    color='blue', alpha=0.1
)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.1f}%')) #Adding "%" to the y-axes


# Costum legend needed since the Confidence interval area doest go automatially on matplotlib
conf_interval_patch = mpatches.Patch(color='blue', alpha=0.1, label='95% Confidence Interval')
handles, labels = ax.get_legend_handles_labels()
handles.append(conf_interval_patch)
plt.legend(handles=handles, loc='upper left')


ax.set_xlabel('Month')
ax.set_ylabel('Average % of Population that had Covid')
ax.set_title('Percentage of Population that got COVID-19 during 2020')

plt.tight_layout()
plt.show()
```

```python
#----------------------------------Importing the data set------------------------------------------------
deaths = pd.read_excel("DataSet/2020_death.xlsx",
                       
                      usecols="I:K",
                      skiprows=3,
                      nrows=8, sheet_name="Death Rates")  

# deaths.head(5)
#--------------------------------------------------------------------------------------------------------


plt.figure(figsize=(9, 4))


sns.barplot(data=deaths, x="Age groups", y="Percentage of all deaths", color="deepskyblue", label="Covid Deaths")

# Overlay the transparent bar that goes up to 100%
total = pd.DataFrame({
    'Age groups': deaths['Age groups'],
    'Total': 100
})
sns.barplot(data=total, x="Age groups", y="Total", color="grey", alpha=0.3, label="Total Deaths")



# Here I am formating the y-axis labels to show percentages
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{int(y)}%'))

plt.title("COVID-19's Impact on Mortality by Age Group 2020")
plt.legend(bbox_to_anchor=(1, 0.8), fancybox=True, shadow=True)
plt.tight_layout()
plt.show()
```

```python
#---------------------------------Importing the sheet from exel file ----------------------------------------------------------------------------
df_1e = pd.read_excel("DataSet/2020_rates.xlsx", sheet_name="1e")

#Fixing the values that seaborn converts them to float and devides them with 100

df_1e["% testing positive for COVID-19"] =df_1e["% testing positive for COVID-19"] * 100 

#-------------------------------------------------------------------------------------------------------------


plt.figure(figsize=(7, 4))
df_1e_sorted = df_1e.sort_values(by="% testing positive for COVID-19", ascending=False)


#Defining palette for better vizualization
palette = sns.light_palette("darkorange", len(df_1e_sorted), reverse=True)





# Creating the barplot for COVID-19 positivity rates
regions = sns.barplot(data=df_1e_sorted, y="Main Areas", x="% testing positive for COVID-19", palette=palette, errcolor="red")
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:.1f}%'))


regions.set_title("The percentage of population tested positive on the main regions in England")
regions.set_ylabel("Regions")
regions.set_xlabel("Percentage Rate")

plt.show()
```

### Step 2: Tracking COVID's Course from Spring 2020 to Early 2023

Next up, I zoomed out to look at the bigger picture of the pandemic's progression.

**How COVID Spread Over Time**:  
I put together a line plot to trace the percentage of population testing postive from April 2020 all the way through to February 2023. It's a snapshot of how the situation has changed over nearly three years.

**Quantifying the Impact**:  
Additionally, to get a sense of the total scale a bar plot was created to quantify the total number of people who got COVID during this period. This visual aims to highlight the extensive reach of the virus.

```python
#Importing the data from the 2023_rates exel file
df_2a = pd.read_excel('DataSet/2023_rates.xlsx', sheet_name='1a')
df_2a['Date Time Period'] = pd.to_datetime(df_2a['Time period'].str.split(' to ').str[1], format='%d %B %Y')
df_2a['Month Year'] = df_2a['Date Time Period'].dt.strftime('%b/%y')


df_2a.head(40)

#-------------------------------------------------------------------------------------------------------------------
plt.figure(figsize=(8, 5))
ax = sns.lineplot(data=df_2a, x='Date Time Period', y='Estimated average % of the population testing positive for COVID-19', label='Estimated Average %')

# Set the date format on the x-axis
date_format = mdates.DateFormatter('%b %y')  # Editing the date, choosing just the month 
ax.xaxis.set_major_formatter(date_format)



plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.1f}%'))





# Costum band for the confeidence interval
plt.fill_between(
    x=df_2a['Date Time Period'],
    y1=df_2a['95% Lower confidence/credible interval for percentage'],
    y2=df_2a['95% Upper confidence/credible interval for percentage'],
    color='blue', alpha=0.1
)

# Costum legend needed since the Confidence interval area doest go automatially on matplotlib
conf_interval_patch = mpatches.Patch(color='blue', alpha=0.1, label='95% Confidence Interval')
handles, labels = ax.get_legend_handles_labels()
handles.append(conf_interval_patch)
plt.legend(handles=handles, loc='upper left')


ax.set_xlabel('Month')
ax.set_ylabel('Avg. Covid Rate (%)')
ax.set_title('Percentage of Population that got COVID-19  2020/2023')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(9, 4.5))
nr_bar = sns.barplot(
    data=df_2a,
    x="Date Time Period",
    y=" Estimated average number of people testing positive for COVID-19",
    color="#FFA500",
    errcolor="grey",
)


#Set a comma between thousands and than change the comma to a point 
nr_bar.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
def custom_formatter(x, pos):
    return '{:,.0f}'.format(x).replace(',', '.')
nr_bar.yaxis.set_major_formatter(ticker.FuncFormatter(custom_formatter))


# Generate the custom date labels for the x-ticks
date_labels = pd.date_range(start="2020-04-01", end="2023-02-28", freq='MS').strftime('%b %y').tolist()
every_other_label = date_labels[::2]  #Slicing and selecting every otehr label
positions = np.linspace(0, len(df_2a['Date Time Period']) - 1, num=len(every_other_label), dtype=int)
nr_bar.set_xticks(positions)

nr_bar.set_xticklabels(every_other_label, rotation=45, ha="right")

#Standart Costumisation
nr_bar.set_xlabel('Month')
nr_bar.set_ylabel('Number of people tested Positive')
nr_bar.set_title('Number of people tested Positive from April 2020 to February 2023')

plt.tight_layout()
plt.show()
```
